```python
from sklearn.preprocessing import LabelEncoder

label_Encoder = LabelEncoder()

df['trans_date_trans_time'] = label_Encoder.fit_transform(df['trans_date_trans_time'])
df['merchant'] = label_Encoder.fit_transform(df['merchant'])
df['category'] = label_Encoder.fit_transform(df['category'])
df['first'] = label_Encoder.fit_transform(df['first'])
df['last'] = label_Encoder.fit_transform(df['last'])
df['gender'] = label_Encoder.fit_transform(df['gender'])
df['street'] = label_Encoder.fit_transform(df['street'])
df['city'] = label_Encoder.fit_transform(df['city'])
df['state'] = label_Encoder.fit_transform(df['state'])
df['job'] = label_Encoder.fit_transform(df['job'])
df['dob'] = label_Encoder.fit_transform(df['dob'])
df['trans_num'] = label_Encoder.fit_transform(df['trans_num'])
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd
```

```python
from imblearn.over_sampling import SMOTE

x = df.drop('is_fraud', axis=1)  # Features
y = df['is_fraud']  # Target variable

# Apply SMOTE with hyperparameter tuning (experiment with different k_neighbors)
smote = SMOTE(k_neighbors=5, random_state=42)  # Set random state for reproducibility
x_resampled, y_resampled = smote.fit_resample(x, y)

# Combine the resampled features and target variable into a DataFrame
balanced_data = pd.concat([x_resampled, y_resampled], axis=1)
```

<div style="background-color: violet; color: black; border-style: solid; border-color: yellow; font-size: 20px; text-align: center"><b><br>Random Forecasting Model<br><br></b></div>

```python
balanced_data.columns
```

```python
features = df[['trans_date_trans_time', 'cc_num', 'merchant', 'category', 'amt',
       'city', 'state', 'lat', 'long', 'trans_num']]
x = features
```

```python
label = df[['is_fraud']]
y = label
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=123)
```

<div style="background-color: violet; color: black; border-style: solid; border-color: yellow; font-size: 20px; text-align: center"><b><br>Model Training<br><br></b></div>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model = RandomForestClassifier(random_state=123)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of Random Forest Classifier Model ==> ", accuracy)
```

```python
from sklearn.metrics import recall_score, precision_score, f1_score

recall = recall_score(y_test, y_pred)
print("Recall of Random Forest Classifier Model ==> ", recall)

precision = precision_score(y_test, y_pred)
print("Precision of Random Forest Classifier Model ==> ", precision)

f1 = f1_score(y_test, y_pred)
print("F1 of Random Forest Classifier Model ==> ", f1)
```

<div style="background-color: violet; color: black; border-style: solid; border-color: yellow; font-size: 20px; text-align: center"><b><br>Saving Model<br><br></b></div>

```python
import pickle
import joblib


# Save the trained model to a file
with open('Random_Forest_Model.pkl', 'wb') as f:
    pickle.dump(model, f)
    
joblib.dump(model, 'Random_Forest_Model.joblib')
```

<div style="background-color: violet; color: black; border-style: solid; border-color: yellow; font-size: 20px; text-align: center"><b><br>Model Testing<br><br></b></div>

```python
x = df[['trans_date_trans_time', 'cc_num', 'merchant', 'category', 'amt', 'city', 'state', 'lat', 'long', 'trans_num']]
y = df['is_fraud']

rf_classifier = RandomForestClassifier(random_state=123)

rf_classifier.fit(x, y)

user_input = {}
for i in x.columns:
    user_input[i] = input("Enter {} : " .format(i))
    
user_df = pd.DataFrame([user_input])

prediction = rf_classifier.predict(user_df)

if prediction[0] == 1:
    print("The Transaction is fraudulent")
else:
    print("The Transaction is not fraudulent")    
```
